```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,2 * np.pi,200)
y=np.sin(x)

fig,ax=plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[1,4,2,3]
fig,ax=plt.subplots()
ax.plot(x,y)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

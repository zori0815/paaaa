```python
import matplotlib.pyplot as plt  
import numpy as np  
  
#   
x = np.arange(5)  
y = np.array([3, 5, 1, 7, 4])  
colors = ['g', 'r', 'c', 'b', 'k']  
  
#   
fig, ax = plt.subplots()  
for i in range(len(x)):  
    ax.bar(x[i], y[i], color=colors[i])  
  
#   
ax.set_xlabel('X Label')  
ax.set_ylabel('Y Label')  
ax.set_title('Bar Chart')  
  
#   
plt.show()
```

```python
import matplotlib.pyplot as plt  
import numpy as np  
  
x = np.linspace(0, 2 * np.pi, 200)  
y = np.sin(x)  
  
fig, ax = plt.subplots()  
ax.plot(x, y)  
  
# scatter  
ax.scatter(x, y, c='r', alpha=0.8)  # calpha  
  
plt.show()
```
